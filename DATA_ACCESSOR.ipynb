# Data Getters

Next we will go over accessor methods. These methods are what help you access the data. I will briefly go over what accessors you can have but you must be sure to have run the first two previous lines of code (verify with the green check mark) in order to run these methods

## GETTER METHODS

These methods allow you to access data from the *food.data* file directly in a visibly appealing way.

### Information Getters
* `get_data_keys()` -> returns a **List** of all available keys in the **Data** dictionary
* `get_list_size()` -> returns an **Integer** of the full size of our dataset

### Indexed Getter Methods
* `get_entry(index: int)` -> returns a **Dictionary** for a single data entry in __food.data__ at the given index.
* `get_entry_description(index: int)` -> returns a **String** for the description of a single entry at the provided index.
* `get_entry_category(index: int)` -> returns a **String** for the category of the food at the provided index.
* `get_nutrient_data_bank_number(index: int)` -> returns an **Integer** for the nutrient data bank number at the provided index.

### Data Getter Methods
* `get_entry_data_dict(index: int)` -> returns a **Dictionary** containing all of the data for the given food entry at that index
* `get_entry_nutrient_score(index: int, nutrient: str)` -> returns an **Integer** containing the nutritional value score given the food entry at the given index. (NOTE: if you want to find out all of the available nutrients run `get_data_keys()`)
* `get_fat_dict_(index: int, datapoint = "")` -> returns the entire fat **Dictionary** at the given index if *datapoint* is not provided because it is optional. If *datapoint* is provided, it will return the integer value at the given index inside of the fat content.
* `get_vitamins_dict(index: int, datapoint = "")` -> returns the entire vitamins **Dictionary** at the given index if *datapoint* is not provided because it is optional. If *datapoint* is provided, it will return the integer value at the given index inside of the viatmins content.

* `get_score_list(index: int)` -> returns a **List** with all of the scores for each point of data at the given index. Runs in O(n) time
